```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,12))

map = Basemap()
map.drawcoastlines()
plt.show()
plt.savefig('coastlines_test.png')
```

```python
fig2, ax = plt.subplots(1,1,figsize=(12,12))
map_ortho = Basemap(projection='ortho', lat_0 = 20, lon_0 = 100)
map_ortho.drawmapboundary(fill_color = 'aqua')
map_ortho.fillcontinents(color='coral',lake_color='aqua')
map_ortho.drawcoastlines()
plt.show()
```

```python
fig2, ax = plt.subplots(1,1,figsize=(12,12))
map_ortho = Basemap(projection='aeqd', lat_0=20, lon_0=100)
map_ortho.drawmapboundary(fill_color = 'aqua')
map_ortho.fillcontinents(color='coral',lake_color='aqua')
map_ortho.drawcoastlines()
plt.show()
```

```python
fig2, ax = plt.subplots(1,1,figsize=(9,9))
map = Basemap(llcrnrlon=3.75,llcrnrlat=39.75,urcrnrlon=4.35,urcrnrlat=40.15, resolution = 'h', epsg=3821)

map.drawmapboundary(fill_color='aqua')
#Fill the continents with the land color
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()
plt.show()
```
