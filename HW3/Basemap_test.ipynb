```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,12))

map = Basemap()
map.drawcoastlines()
plt.show()
plt.savefig('coastlines_test.png')
```

```python
fig2, ax = plt.subplots(1,1,figsize=(12,12))
map_ortho = Basemap(projection='ortho', lat_0 = 20, lon_0 = 100)
map_ortho.drawmapboundary(fill_color = 'aqua')
map_ortho.fillcontinents(color='coral',lake_color='aqua')
map_ortho.drawcoastlines()
plt.show()
```

```python
fig3, ax = plt.subplots(1,1,figsize=(12,12))
map_ortho = Basemap(projection='aeqd', lat_0=20, lon_0=100)
map_ortho.drawmapboundary(fill_color = 'aqua')
map_ortho.fillcontinents(color='coral',lake_color='aqua')
map_ortho.drawcoastlines()
plt.show()
```

```python
fig4, ax = plt.subplots(1,1,figsize=(12,12))
map = Basemap(llcrnrlon=73.62,llcrnrlat=16.7,urcrnrlon=134.77,urcrnrlat=53.56, resolution = 'h', epsg=4490)

map.drawmapboundary(fill_color='aqua')
#Fill the continents with the land color
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()
plt.show()
```

```python
fig3, ax = plt.subplots(1,1,figsize=(9,9))
map = Basemap(projection='aeqd',
              lon_0 = 110,
              lat_0 = 40,
              width = 7000000,
              height = 5000000)

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()

for i in range(0, 10000000, 1000000):
    map.plot(i, i, marker='*',color='k')

plt.show()
```
