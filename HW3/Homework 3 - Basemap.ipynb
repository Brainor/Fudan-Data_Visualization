# 1. 请使用python (+Basemap)可视化完整的中国领土的地图（陆地和海洋）。使用至少3种不同的投影方法，必须包括conformal, equal area, equal distance (Azimuthal) projections。在使用Basemap函数时，需要注释说明调用的函数中的各个参数的意义

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,9))
China_map = Basemap(llcrnrlon=90, llcrnrlat=2, urcrnrlon=140, urcrnrlat=60, 
                    projection='lcc', lat_0 = 42.5,lon_0=120)


China_map.drawcoastlines()
# China_map.bluemarble()
China_map.drawcountries(linewidth=1)
plt.show()
```
